```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir 'Fecha de Admisión' a datetime si no lo es
# Convierte la columna 'Fecha de Admisión' a un formato datetime, si no lo está.
df['Fecha de Admisión'] = pd.to_datetime(df['Fecha de Admisión'])

# Extraer el mes de la columna 'Fecha de Admisión'
# Crea una nueva columna 'Mes de Admisión' con solo el mes extraído de 'Fecha de Admisión'
df['Mes de Admisión'] = df['Fecha de Admisión'].dt.month

# Agrupar por mes y contar admisiones
# Agrupa el DataFrame por 'Mes de Admisión' y cuenta la cantidad de registros por mes
admissions_by_month = df.groupby('Mes de Admisión').size()

# Calcular la media de admisiones por mes
# Calcula el promedio de las admisiones por mes
media_admisiones_por_mes = admissions_by_month.mean()

# Crear el gráfico
plt.figure(figsize=(10, 6))  # Establece el tamaño de la figura

# Línea con marcador y color personalizado
# Grafica las admisiones por mes con un marcador circular, color 'deepskyblue', y líneas
plt.plot(admissions_by_month.index, admissions_by_month.values, marker='o', color='deepskyblue', linestyle='-', linewidth=2, markersize=8)

# Personalización de etiquetas y título
# Personaliza las etiquetas de los ejes y el título del gráfico
plt.xlabel('Mes de Admisión', fontsize=12, fontweight='bold', color='darkslategray')
plt.ylabel('Número de Admisiones', fontsize=12, fontweight='bold', color='darkslategray')
plt.title('Gráfico de Admisiones por Mes', fontsize=14, fontweight='bold', color='darkblue')

# Configuración de los ticks en el eje x
# Configura las marcas en el eje x para mostrar los números de los meses (1-12) con una rotación de 45 grados
plt.xticks(range(1, 13), fontsize=11, rotation=45, color='dimgray')
plt.yticks(fontsize=11, color='dimgray')

# Agregar etiquetas a los puntos con mayor distancia
# Añade las etiquetas de las cantidades de admisiones a cada punto en el gráfico
for x, y in zip(admissions_by_month.index, admissions_by_month.values):
    plt.text(x, y, str(y), fontsize=10, ha='center', va='bottom', color='darkblue', fontweight='bold')

# Estilo de la cuadrícula
# Agrega una cuadrícula al gráfico con estilo de líneas discontinuas y color gris
plt.grid(True, linestyle='--', alpha=0.5, color='gray')

# Leyenda
# Agrega una leyenda al gráfico
plt.legend(['Admisiones'], fontsize=12, loc='upper left')

# Ajuste final de diseño
# Ajusta el diseño para evitar el corte de los elementos del gráfico
plt.tight_layout()

# Mostrar gráfico
# Muestra el gráfico resultante
plt.show()

# Imprimir la media de admisiones por mes
# Imprime en consola la media de admisiones por mes
print(f"La media de admisiones por mes es: {media_admisiones_por_mes}")
```

```python
# Define the columns for the pie charts
columns = ['Fumador', 'Alcohol', 'Diabetes Mellitus', 'Hipertensión']

# Crear una figura con subgráficos (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # Tamaño ajustado para visibilidad

# Iterar sobre las columnas y las posiciones de los subgráficos
for col, ax in zip(columns, axes.flatten()):
    # Count the occurrences of 0 and 1
    value_counts = df[col].value_counts()

    # Calculate percentages
    percentages = (value_counts / value_counts.sum()) * 100

    # Prepare labels for the pie chart, including total counts and percentages
    labels = [f'{label} ({count}) - {percent:.1f}%'
              for label, count, percent in zip(value_counts.index, value_counts.values, percentages)]

    # Crear el gráfico de pastel en el subgráfico correspondiente
    ax.pie(value_counts, labels=labels, autopct=None, startangle=90,
           colors=sns.color_palette('pastel'))  # Usar paleta de colores pastel

    # Título del subgráfico
    ax.set_title(f'Distribución de {col}', fontsize=14, fontweight='bold')

    # Asegurar que los gráficos sean círculos
    ax.axis('equal')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar todos los gráficos en una sola figura
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Columnas para los gráficos
columns = [
    'Enfermedad Coronaria Arterial',
    'Miocardiopatía', 'Enfermedad Renal Crónica', 'Anemia Severa', 'Anemia',
    'Angina Estable', 'Síndrome Coronario Agudo', 'Infarto de Miocardio(STEMI)',
    'Dolor Torácico Atípico', 'Insuficiencia Cardíaca',
    'Insuficiencia Cardíaca con Fracción de Eyección Reducida',
    'Insuficiencia Cardíaca con Fracción de Eyección Normal',
    'Enfermedad Valvular Cardíaca', 'Bloqueo Cardíaco Completo',
    'Síndrome del Nodo Sinusal Enfermo', 'Lesión Renal Aguda',
    'Accidente Cerebrovascular Isquémico', 'Accidente Cerebrovascular Hemorrágico',
    'Fibrilación Auricular', 'Taquicardia Ventricular',
    'Taquicardia Supraventricular Paroxística', 'Cardiopatía Congénita',
    'Infección del Tracto Urinario', 'Síncope Neurocardiogénico', 'Ortostático',
    'Endocarditis Infecciosa', 'Trombosis Venosa Profunda', 'Choque Cardiogénico',
    'Shock', 'Embolia Pulmonar', 'Infección Torácica'
]

# Configurar filas y columnas dinámicamente
n_cols = 3  # Número de columnas deseado
n_rows = math.ceil(len(columns) / n_cols)  # Calcula el número de filas necesarias

# Crear figura con subgráficos
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))  # Tamaño dinámico
axes = axes.flatten()  # Aplanar los ejes para iterar más fácilmente

# Iterar sobre columnas y subgráficos
for i, col in enumerate(columns):
    # Verificar que la columna exista en el DataFrame
    if col in df.columns:
        # Contar valores
        value_counts = df[col].value_counts()

        # Calcular porcentajes
        percentages = (value_counts / value_counts.sum()) * 100

        # Etiquetas para el gráfico
        labels = [f'{label} ({count}) - {percent:.1f}%'
                  for label, count, percent in zip(value_counts.index, value_counts.values, percentages)]

        # Asignar colores: Rojo para "Sí" (1) y Verde para "No" (0)
        colors = ['red' if label == 1 else 'green' for label in value_counts.index]

        # Crear gráfico de pastel
        axes[i].pie(value_counts, labels=labels, autopct=None, startangle=90,
                    colors=colors)
        axes[i].set_title(f'{col}', fontsize=12, fontweight='bold')
        axes[i].axis('equal')  # Asegurar forma de círculo
    else:
        axes[i].axis('off')  # Desactivar gráfico si no hay datos

# Desactivar gráficos extra si hay más ejes que columnas
for j in range(len(columns), len(axes)):
    axes[j].axis('off')

# Ajustar espaciado
plt.tight_layout()
plt.subplots_adjust(hspace=0.4)  # Espaciado horizontal entre gráficos

# Mostrar gráficos
plt.show()
```
